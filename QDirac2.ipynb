```python
def type_validate(x):
    if isinstance(x, int):
        return [x]
    elif isinstance(x, list):
        if all([isinstance(_, int) for _ in x]):
            return x
        else:
            raise TypeError("All elements of input list (x) must be integer type")
    else:
        raise ValueError("Only Integer and List of Integers allowed")
```

```python
def apply_gate(gate, vector_dict, control, target):
    """
    Waring: User defined/custom function: 
    ***always run from numpy import allclose, array***
    before running/using this method. also define 
    KET_1 = array([[0], [1]]).
    ***NOT TO BE USED OUT OF SCOPE OF THIS LIBRARY***
    ##
    ## gate: 2x2 numpy array;
    ## vector_dict -> typically {<int>: numpy-array-2x1}
    ## control -> int -> must be within key set of vector_dict
    ## target -> int -> must be within key set of vector_dict
    """
    if allclose(vector_dict[control], KET_1):
        vector_dict[target] = matmul(gate, vector_dict[target])
        
    return vector_dict
```

```python
def two_qubit_operation(n, gate, source_qubit, target_qubit):
    """
    """
    In =  eye(2**n)
    try:
        Oracle = list()
        for i_n in In:
            r=int( "".join([str(int(x)) for x in i_n.tolist()][::-1]), 2)
            ik = int(np.log2(r))
            xik = bin(ik)[2:].zfill(n)
            bits_dict = {i: int(x_) for i, x_ in enumerate(xik)}
            basis_ket_dict = {k: Z_BASIS_KET_DICT[v] for k, v in bits_dict.items()}
            plain_ket = reduce(kron, list(basis_ket_dict.values()))
            
            gate_operated_dict = apply_gate(gate, basis_ket_dict, source_qubit, target_qubit)
            #z_basis_ket_dict ={k: Z_BASIS_KET_DICT[v] for k, v in bits_dict.items()}
            
            x_operated_ket = reduce(kron, list(gate_operated_dict.values()))
            Oracle.append(matmul(plain_ket, x_operated_ket.T))

        return sum(Oracle)
      
    except:
        raise Exception("TBA")
```

```python
def apply_ccnot(gate, vector_dict:dict, controls:list, target:int):
    """
    Waring: User defined/custom function: 
    ***always run from numpy import allclose, array***
    before running/using this method. also define 
    KET_1 = array([[0], [1]]).
    ***NOT TO BE USED OUT OF SCOPE OF THIS LIBRARY***
    ##
    ## gate: 2x2 numpy array;
    ## vector_dict -> typically {<int>: numpy-array-2x1}
    ## control -> int -> must be within key set of vector_dict
    ## target -> int -> must be within key set of vector_dict
    """
    conditions_match_list = [allclose(vector_dict[k] , KET_1) for k in controls]
    if all(conditions_match_list):
        vector_dict[target] = matmul(gate, vector_dict[target])

    return vector_dict
```

```python
def ccnot_operation(n, gate, control_qubits, target_qubit):
    In =  eye(2**n)
    try:
        Oracle = list()
        for i_n in In:
            r=int( "".join([str(int(x)) for x in i_n.tolist()][::-1]), 2)
            ik = int(np.log2(r))
            xik = bin(ik)[2:].zfill(n)
            bits_dict = {i: int(x_) for i, x_ in enumerate(xik)}
            basis_ket_dict = {k: Z_BASIS_KET_DICT[v] for k, v in bits_dict.items()}
            plain_ket = reduce(kron, list(basis_ket_dict.values()))
            
            gate_operated_dict = apply_ccnot(gate, basis_ket_dict, control_qubits, target_qubit)
            #z_basis_ket_dict ={k: Z_BASIS_KET_DICT[v] for k, v in bits_dict.items()}
            
            uf_operated_ket = reduce(kron, list(gate_operated_dict.values()))
            Oracle.append(matmul(plain_ket, uf_operated_ket.T))

        return sum(Oracle)
      
    except:
        raise Exception("TBA")
```

```python
class QuantumCircuit():
    def __init__(self, num_of_qubits):
        self.n = num_of_qubits
        self._STATE_VECTOR = None
        print("Initialize by <instance>.init()")
        
    def init(self):
        self._STATE_VECTOR = reduce(kron, [KET_0 for i in range(self.n)])
        init_circuit(self.n)
        Wires()
    
    def state_vector(self):
        """
        Returns Global State of all qubits only
        """
        return self._STATE_VECTOR
        


    def _single_qubit_operation(self, on_qubits, gate):
        Uf =  list()
        if isinstance(on_qubits, int):
            Uf = [gate if i==on_qubits else eye(2) for i in range(0, self.n) ]
        
            
        if isinstance(on_qubits, list):
            Uf = [gate if i in on_qubits else eye(2) for i in range(0, self.n) ]
           
            
        self._STATE_VECTOR = matmul(reduce(kron, Uf), self._STATE_VECTOR)

            
     

    ##### Quantum Gates or Unitary operations
    def X(self, on_qubits):
        self._single_qubit_operation(on_qubits, X)
        generic_single_gate({i:"X" for i in type_validate(on_qubits)})
        

    def Y(self, on_qubits):
        self._single_qubit_operation(on_qubits, Y)
        generic_single_gate({i:"Y" for i in type_validate(on_qubits)})
    
    def Z(self, on_qubits):
        self._single_qubit_operation(on_qubits, Z)
        conv2list = lambda x: [x] if isinstance(x, int) else x
        generic_single_gate({i:"Z" for i in type_validate(on_qubits)})

    def H(self, on_qubits):
        self._single_qubit_operation(on_qubits, H)
        generic_single_gate({i:"H" for i in type_validate(on_qubits)})
        

    def CX(self, source_qubit:int, target_qubit:int):
        """
        CX is a Two qubit Universal Gate. It is also called Non-local gate( or :At a distance"from Einstein's(EPR Paradox based - Spooky action 'at distance').
            It is a Unitary Gate.
        """
        qubit_gap = source_qubit - target_qubit
        if qubit_gap ==0:
            raise Exception("CNOT Doesn't operate on same same/single Qubit: Source Qubit and Target index can't be the same same")
            
        if source_qubit < self.n and target_qubit < self.n and qubit_gap < 0:
            Uf = two_qubit_operation(self.n, X, source_qubit, target_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)

        else:
            Uf = libs2.two_qubit_operation(self.n, X, target_qubit, source_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)

        cnot_gate(source_qubit, target_qubit)
            

    def CZ(self, source_qubit:int, target_qubit:int):
        """
        CZ is a Two qubit Universal Gate. It is also called Non-local gate( or :At a distance"from Einstein's(EPR Paradox based - Spooky action 'at distance').
            It is a Unitary Gate.
        """
        qubit_gap = source_qubit - target_qubit
        if qubit_gap ==0:
            raise Exception("CNOT Doesn't operate on same same/single Qubit: Source Qubit and Target index can't be the same same")
            
        if source_qubit < self.n and target_qubit < self.n and qubit_gap < 0:
            Uf = two_qubit_operation(self.n, Z, source_qubit, target_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)

        else:
            Uf = libs2.two_qubit_operation(self.n, Z, target_qubit, source_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)

        cz_gate(source_qubit, target_qubit)
        

    def CY(self, source_qubit:int, target_qubit:int):
        """
        CY is a Two qubit Universal Gate. It is also called Non-local gate( or :At a distance"from Einstein's(EPR Paradox based - Spooky action 'at distance').
            It is a Unitary Gate.
        """
        qubit_gap = source_qubit - target_qubit
        if qubit_gap ==0:
            raise Exception("CNOT Doesn't operate on same same/single Qubit: Source Qubit and Target index can't be the same same")
            
        if source_qubit < self.n and target_qubit < self.n and qubit_gap < 0:
            Uf = two_qubit_operation(self.n, Y, source_qubit, target_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)

        else:
            Uf = libs2.two_qubit_operation(self.n, Y, target_qubit, source_qubit)
            self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)


    def CCX(self, control_qubits:list, target_qubit):
        if isinstance(target_qubit, int):
            if all([(isinstance(x, int) and 0 >= x < self.n) for x in control_qubits]):
                Uf = ccnot_operation(self.n, X, control_qubits, target_qubit)
                self._STATE_VECTOR = matmul(Uf, self._STATE_VECTOR)
                

              

    def SWAP(self, source_qubit:int,  target_qubit:int):
        Uf_1 = two_qubit_operation(self.n, X, source_qubit, target_qubit)
        Uf_2 = two_qubit_operation(self.n, X, target_qubit, source_qubit)
        self._STATE_VECTOR = matmul(Uf_1@Uf_2@Uf_1, self._STATE_VECTOR)
        swap_gate(source_qubit, target_qubit)


    def measure(self, basis="Z"):
        bra_vector_n = reduce(kron, [KET_1.T for i in range(self.n)])
        measured_result = matmul(bra_vector_n, self._STATE_VECTOR)
        measurement_gate()
        global SVG_EXCHANGE_OBJECT
        SVG_EXCHANGE_OBJECT['measurement-applied'] = True
        return measured_result


    def draw(self):
        global SVG_EXCHANGE_OBJECT
        draw_svg()
        if SVG_EXCHANGE_OBJECT['measurement-applied']:
            del SVG_EXCHANGE_OBJECT

        
        
```
